```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

NOW WORKING IN CAST COLUMN
--i want to concider only top 3 

```python
movies['cast']
```

```python
import ast

def convert3(obj):
    L = []
    count = 0
    try:
        data = ast.literal_eval(obj)
        for i in data:
            if count != 3:
                L.append(i['name'])
                count += 1
            else:
                break
    except (ValueError, SyntaxError, TypeError):
        return []
    return L
    
```

```python
movies['cast'] = movies['cast'].apply(lambda x: convert3(x) if isinstance(x, str) else [])
```

```python
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

CONVERTING TEXT INTO NUMERIC FORM(VECTORIZATION) --(BAGS OF WORDS)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags'])
```

```python
vectors.toarray()
```

```python
cv.get_feature_names_out()
```

HERE WE DON'T WANT TO EXTRA REPETATION IN A DATA JUST REMOVE THE EXTRA

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

this is example--

```python
ps.stem('loving')
```

NOW WE ARE USING STEM FUNCTION IN "TAGS" COLUMN

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

NOW MAKINNG A RECOMMEND SYSTEM

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances= similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
